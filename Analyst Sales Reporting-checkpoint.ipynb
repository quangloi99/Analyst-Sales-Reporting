# Task 1: Import packages and Load data

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
# filename = input('D:\Fiance\Sales Reporting -20211213T005009Z-001\Sales Reporting\sales2019_1.csv ')
Paths = 'D:/Fiance/Sales Reporting -20211213T005009Z-001/Sales Reporting/'
df = pd.read_csv(Paths +'sales2019_1.csv ')
df.head()
```

# Task 2: Clean and Preprocess data

2.1: Merge 12-month data

```python
path = "D:\Fiance\Sales Reporting -20211213T005009Z-001\Sales Reporting"
# Paths = D:\Fiance\Sales Reporting -20211213T005009Z-001\Sales Reporting\"
frames =[]
for file in os.listdir(path):
    if file.endswith('.csv'):
        path_di = Paths + file
        df1 = pd.read_csv(path_di)
        frames.append(df1)
        result = pd.concat(frames)
result 
result.to_csv('sale_2019.csv',index = False)
```

Check Datatype of series

```python
df = result
df.info()
```

2.2: Add 'Month' column

```python

df["Month"] = ''
df["Month"]= df['Order Date'].str[0:2]
df.head()
```

```python
print(set(df["Month"]))
```

2.3 GET RID OF VALUES 'NAN' AND 'OR'

```python
df = df[df["Month"] != 'Or']
df.head()
```

```python
df = df.dropna(how = 'all')
print(set(df["Month"]))
```

# TASK 3 REPORTING

CONVERT TO NUMBERIC

```python
df['Price Each']= pd.to_numeric(df['Price Each'], downcast="float")
df['Quantity Ordered']= pd.to_numeric(df['Quantity Ordered'], downcast="integer")
```

Check datatype

```python
print(df['Price Each'].dtype)
print(df['Quantity Ordered'].dtype)
```

```python
df['Sales'] = df['Price Each']*df['Quantity Ordered']
df.head()
```

MOVING COLUMNS

```python
moving_col = df.pop('Sales')
df.insert(4,'Sales',moving_col)
df.head()
```

```python
df.groupby('Month').sum()['Sales']
```

```python
 sale_value  = df.groupby('Month').sum()['Sales']
sale_value.max()
```

# Tháng nào bán được hàng nhiều nhất?

```python
month = range(1,13)
plt.bar(x = month ,height= sale_value)
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.title("Revenue by Month")
plt.grid()
plt.show();
```

#    Thành phố nào có doanh số bán hàng cao nhất?

```python
df.head()
```

```python
 adress  = lambda sample:sample.split(',')[1]
```

```python
df['City'] = df['Purchase Address'].apply(adress)
df.head()
```

```python
df.groupby('City').sum()['Sales']
```

```python
Sales_City =df.groupby('City').sum()['Sales']
Sales_City.max()
```

=> San Francisco là thành phố có doanh số bán hàng cao nhất

```python
city =[city for city, v in Sales_City.items()]
    
plt.bar(city,Sales_City)
plt.xticks(city,rotation = 90, size = 8)
plt.xlabel("City")
plt.ylabel("Revenue")
plt.title(" Revenue by City")
plt.show()
```

#  Thời gian nào nên chạy quảng cáo để tối ưu hóa việc bán hàng?

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
print(df['Order Date'].dtype)
```

```python
df["hours"] = df['Order Date'].dt.hour
df.head()
```

```python
sale_hour =  df.groupby('hours').count()['Sales']
Hour =[hour for hour, sales in sale_hour.items()]   
plt.plot(Hour,sale_hour, color = 'blue' )
plt.grid()
plt.xticks(Hour,rotation = 90, size = 8)
plt.xlabel("Hours")
plt.ylabel("Revenue")
plt.title(" Revenue by Hours")
plt.show()
```

```python
#  Hai khung giờ khách hàng mua nhiều nhất là giờ nghỉ trưa(11-13h) và buổi tối(16-20h)
#  Do đó nên chạy quảng cáo trước khung giờ này khoảng từ 30'-1h để tối ưu hóa việc bán hàng cho doanh nghiệp
```

# Sản phẩm nào được bán nhiều nhất?

```python
df.groupby(df['Product']).sum()['Quantity Ordered']
```

```python
quantity = df.groupby(df['Product']).sum()['Quantity Ordered']
name =[name for name, quantity in quantity.items()]   
plt.bar(name,quantity)
plt.xticks(Hour,rotation = 90)
plt.xlabel("Item")
plt.ylabel("Revenue")
plt.title(" Revenue by Item")
plt.show()
```

```python
# hai sản phẩm bán được nhiều nhất là: AA Batteries và AAA Batteries
# các yếu tố ảnh hưởng: giá sản phẩm
```

```python
price_product = df.groupby(df['Product']).mean()['Price Each']
```

```python
x = name
y1 = quantity
y2 = price_product

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(x, y1)
ax2.plot(x, y2, color = 'r')
ax1.set_xticklabels(name,rotation = 90)
ax1.set_xlabel('Items',rotation = 90)
ax1.set_ylabel('Sales',color = 'b')
ax2.set_ylabel('Price', color='r')

plt.show();
```

```python
#  Giá sản phẩm tỷ lệ với số lượng sản phẩm bán ra
#  các sản phẩm có giá càng thấp thì sẽ bán được nhiều hơn đáng kể so với các sản phẩm giá cao
```
